## 11. Contract


```python
check_errors(fifa, 11)
```

Creating three new columns to hold the start of the contract, the end of the contract and the contract length.

```python
def get_contract_details(contract):
    if contract == 'Free' or 'On Loan' in contract:
        start_date = np.nan
        end_date = np.nan
        contract_length = 0
    else:
        start_date,  end_date= contract.split(' ~ ')
        start_date = int(start_date[:4])
        end_date = int(end_date[:4])
        contract_length = end_date - start_date
    return start_date, end_date, contract_length
```

```python
new_cols = ['Contract Start', 'Contract End', 'Contract Length']
new_data = fifa['Contract'].apply(lambda x: pd.Series(get_contract_details(x), dtype = 'float'))
```

```python
for i, columns in enumerate(new_cols):
    fifa.insert(loc = fifa.columns.get_loc('Contract')+1+i, column = columns, value = new_data[i])
```

```python
#Viewing a random subset
fifa.sample(3)
```

Lets also insert another column to hold the contract status of the player.

```python
def get_contract_status(contract):
    if contract == 'Free':
        status = 'Free Agent'
    elif 'On Loan' in contract:
        status = 'On Loan'
    else:
        status = 'Contracted'
    return status
    
```

```python
stats = fifa['Contract'].apply(lambda x: pd.Series(get_contract_status(x)))
fifa.insert(loc = fifa.columns.get_loc('Contract Length')+1, column = 'Contract Status', value = stats)
```

## 13. Height


```python
check_errors(fifa, 13)
```

Some of the values in the height column are in centimetres while others are in feet, the column values will be changed to centimetres

```python
def normalize_height(height):
    if 'cm' in height:
        return int(height.replace('cm', ''))
    else:
        feet, inches = height.split("'")
        return round(30.48 * int(feet) + int(inches.strip('"')))
    
```

```python
 fifa['Height']= fifa['Height'].apply(normalize_height)
```

The height column has been converted to centimetres successfully.

## 14. Weight


```python
check_errors(fifa, 14)
```

```python
def convert_weight(weight):
    if 'kg' in weight:
        return int(weight.replace('kg', ''))
    else:
        return round(0.454*(int(weight.replace('lbs', ''))))
```

```python
fifa['Weight'] = fifa['Weight'].apply(convert_weight)
```

## 20. Value

The value format is in an object format which is basically a string, we will need this column to be in an integer format. 

A function will be created to convert this string formatted column to an integer format. This function will also aid in formatting the wages and release clause columns

```python
def get_monetary_value(value):
    if 'M' in value:
        return int(float(value.replace('M', '')[1:]) * 1000000)
    elif 'K' in value:
        return int(float(value.replace('K', '')[1:]) * 1000)
    else:
        return int(value[1:])
        
    
```

```python
fifa['Value'] = fifa['Value'].apply(get_monetary_value)
fifa.Value.unique()
```
